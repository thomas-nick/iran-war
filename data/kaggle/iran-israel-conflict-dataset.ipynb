```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
"Country":["Iran","Iran","Iran","Iran","Israel","Israel","Israel","Israel"],
"Date":["2026-02-28","2026-03-01","2026-03-03","2026-03-05",
        "2026-02-28","2026-03-01","2026-03-03","2026-03-05"],
"Missile_Attacks":[120,150,180,200,30,40,50,60],
"Drone_Attacks":[80,90,100,120,20,25,30,35],
"Deaths":[350,420,480,520,5,7,9,12],
"Injuries":[2000,2500,3000,3500,120,180,240,310],
"Civilian_Deaths":[70,90,110,130,4,6,8,10],
"Military_Deaths":[280,330,370,390,1,1,1,2],
"Infrastructure_Damage":[45,60,75,85,10,12,15,18]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.loc[2,"Injuries"] = np.nan
df.loc[5,"Drone_Attacks"] = np.nan

print(df)
```

```python
print(df.isnull().sum())
```

```python
df["Deaths"].fillna(df["Deaths"].mean(), inplace=False)
```

```python
df["Drone_Attacks"].fillna(df["Drone_Attacks"].median(), inplace=False)

print(df)
```

```python
# pandas plot 
df.groupby("Country")["Deaths"].sum().plot(kind="bar")
plt.title("Total Deaths by Country")
plt.xlabel("Country")
plt.ylabel("Deaths")
plt.show()
```

```python
df.plot(x="Date", y="Missile_Attacks", kind="line")

plt.title("Missile Attacks Trend")
plt.show()
```

```python
sns.scatterplot(x="Missile_Attacks", y="Deaths", hue="Country", data=df)

plt.title("Missile Attacks vs Deaths")
plt.show()
```

```python
sns.boxplot(x="Country", y="Injuries", data=df)

plt.title("Injuries Distribution by Country")
plt.show()
```

```python
sns.histplot(df["Deaths"], kde=True)

plt.title("Deaths Distribution")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```
